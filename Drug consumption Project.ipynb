## Machine Learning

```python
from sklearn.preprocessing import StandardScaler,label_binarize
from sklearn.svm import LinearSVC
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix,accuracy_score,precision_score,recall_score,f1_score
import warnings
warnings.filterwarnings("ignore")
```

```python
Durg_data2 = Durg_data.copy()
for D in DC :
    Durg_data2.loc[Durg_data[D]==0, D]= 0
    Durg_data2.loc[Durg_data[D]==1, D]= 0
    Durg_data2.loc[Durg_data[D]==2, D]= 1
    Durg_data2.loc[Durg_data[D]==3, D]= 1
    Durg_data2.loc[Durg_data[D]==4, D]= 1
    Durg_data2.loc[Durg_data[D]==5, D]= 1
    Durg_data2.loc[Durg_data[D]==6, D]= 1
```

```python
MD = Durg_data2.drop(['Age', 'Gender', 'Education','Country','Ethnicity','Alcohol','Amphet','Amyl','Benzos','Caffeine','Chocolate','Cocaine','Crack','Ecstasy','Heroin','Ketamine','Legalh','LSD','Meth','Mushrooms','Nicotine','Semeron','VSA'], axis=1)
target = MD['Cannabis']
features = MD.drop('Cannabis', axis=1)

xtrain, xtest, ytrain, ytest = train_test_split(features, target, test_size=0.20, shuffle=True, stratify=target)

```

```python
def Confusion_Matrix(ytest, p, accuracy):
    cm = confusion_matrix(ytest, p)
    accuracy = accuracy_score(ytest, p)
    plt.matshow  (cm)
    plt.xlabel   ('Predicted Labels')
    plt.ylabel   ('True Labels')
    plt.suptitle ('Confusion matrix of %s (Accuracy of %.2f) '%(mn,accuracy))
    plt.colorbar()
    plt.show ()
    
```

### Logistic Regression

```python
from sklearn.linear_model import LogisticRegression

LG = LogisticRegression().fit(xtrain, ytrain)

P = LG.predict(xtest)

print(classification_report(ytest, P))

R1 = accuracy_score(ytest, P)
R1


```

```python
mn = 'Logistic Regression'
Confusion_Matrix(ytest, P ,mn)
```

### KNN

```python
from sklearn.neighbors import KNeighborsClassifier

KNN = KNeighborsClassifier().fit(xtrain, ytrain)
p = KNN.predict(xtest)
print(classification_report(ytest, p))

R2 = accuracy_score(ytest, p)
R1


```

```python
mn = 'KNearest Neighbors'
Confusion_Matrix(ytest, p ,mn)
```

### Random Forest


```python
from sklearn.ensemble import RandomForestClassifier

RD = RandomForestClassifier().fit(xtrain, ytrain)
p = RD.predict(xtest)
print(classification_report(ytest, p))

R3 = accuracy_score(ytest, p)
R3


```

```python
mn = 'Random Forest'
Confusion_Matrix(ytest, p,mn)
```
